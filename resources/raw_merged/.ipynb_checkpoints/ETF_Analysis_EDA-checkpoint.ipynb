```python
etf_trade_data = df[['Ticker', 
                    'Open', 
                    'High', 
                    'Low', 
                    'Close', 
                    'Volume', 
                    'Dividends', 
                    'Stock Splits']]

commod_df = df['WTI_Crude_Oil_Price']

composit_ind_df = df[['Chicago_Fed_National_Activity_Index',
                      'Chicago_Fed_National_Financial_Conditions']]

cred_spread_df = df[['Moodys_BAA_Corporate_Bond_Yield',
                     'High_Yield_OAS',
                     'BBB_Corporate_Spread']]

expectations_df = df[['U_Mich_Consumer_Sentiment',
                      'U_Mich_Inflation_Expectations']]

housing_df = df[['New_Private_Housing_Units_Authorized',
                 'Housing_Starts:_Total',
                 '30-Year_Fixed_Mortgage_Rate']]

interest_rate_df = df[['10-Year_Breakeven_Inflation', '5-Year_Breakeven_Inflation',
                       'Effective_Fed_Funds_Rate', '10-Year_TIPS_Yield',
                       '10Y-2Y_Treasury_Spread', '10Y-3M_Treasury_Spread',
                       '1-Month_Treasury_Bill','10-Year_Treasury_Yield',
                       '2-Year_Treasury_Yield']]

labour_market_df = df[['Nonfarm_Payrolls', 'Job_Openings_(JOLTS)',
                       'Civilian_Labor_Force_Participation', 'Natural_Rate__of_Unemployment_(est.)',
                       'Unemployment_Rate']]

markets_df = df[['Equity_Vol_Tracker:_Broad Macro',
                 'CBOE_Volatility_Index (VIX)',
                 'Equity_Mkt_Vol:_Macro_News_Tracker',
                 'Equity_Market_Uncertainty_Index']]

money_credit_df = df[['Commercial_&_Industrial_Loans', 'M2_Money_Stock',
                      'Monetary_Base', 'ON_RRP_(Reverse_Repo)_Balance']]

real_activity_df = df[['Capacity_Utilization:_Manufacturing', 'Leading_Index_for_US',
                       'Personal_Consumption_Expenditures', 'Industrial_Production_Index',
                       'CPI_Inflation', 'GDP',
                       'Real_Gross_Domestic_Product', 'NBER_Recession_Indicator',
                       'Recession_Probabilities']]

prices_df = df[['Core_CPI_(ex_Food/Energy)',
                'Average_price_of_electricity',
                'Average_price_of_eggs',
                'Producer_Price_Index']]
```

```python
df_all = [etf_trade_data, commod_df, composit_ind_df, cred_spread_df, expectations_df,
          housing_df, interest_rate_df, labour_market_df, markets_df,
          money_credit_df, real_activity_df, prices_df]

df_names = ['ETF Trade Data', 'Commodities', 'Composite Index', 'Credit Spread', 'Expectations',
            'Housing', 'Interest Rates', 'Labor Market', 'Markets',
            'Money/Credit', 'Real Activity', 'Prices']
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(14,4,figsize=(20,30)) #creates a figure and 64 sub-plots

axes=axes.ravel() #flattens 2D array

i=0 #initialises the sub-plot counter

for name,df_group in zip(df_names,df_all): #pairing the above names (df_names) vs counter-part (df_all)
    
    if 'Series' in str(type(df_group)): #for single columns (there's only one to deal with here - commodities)
        
        axes[i].hist(df_group.dropna(), #Drops any NaNs found
                     bins=30, #30 bars
                     alpha=.8, #transparency level (set at 80%)
                     color='teal') #bar colours
        axes[i].set_title(name) #sets just title to feature name
        i+=1 #onto the next sub-plot by adding 1 to the counter
    
    else:
        
        for col in df_group.columns:  # looping through all multi-column
            if i<len(axes): #checks if it exceeds 48 sub-plots 
                axes[i].hist(df_group[col].dropna(),
                             bins=30, #as above
                             alpha=.8, #as above
                             color='blue') #as above
                axes[i].set_title(f'{name}: {col}') #sets group title:group member column here
                i+=1 #as above

plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number')
corr = numeric_cols.corr(method='pearson')
```

```python
corr
```

```python
# Top correlations with ETF Close Price
close_corr = corr['Close'].sort_values(ascending=False)
print("\nCorrelation of numeric features with Close:\n")
print(close_corr * 100)
```

```python
import numpy as np
# Find pairs of features with high absolute correlation

# Flatten the correlation matrix into pairs
corr_pairs = (
    corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    .stack()
    .reset_index()
)
corr_pairs.columns = ["feature_1", "feature_2", "correlation"]

# Filter pairs with |corr| above a threshold
threshold = 0.9
high_corr_pairs = corr_pairs[abs(corr_pairs["correlation"]) >= threshold] \
                      .sort_values(by="correlation", ascending=False)

print(f"Highly correlated feature pairs (|corr| >= {threshold}):\n")
print(high_corr_pairs.head(20))
```

```python
import numpy as np
"""
1. taking out the negative correlations & making them ALL positive - for sake of analysis
2. using a mask to make a new list from above table and selecting features just once without their duplicates
   (avoiding duplicate pairs - like we did for the pairplot graphs below)
3. all duplicate values set to 'NaN'; 
4. stacks remaining values into long format; index reset
5. the remaining values' columns are renamed
6. Apply threshold and remove duplicates/self-pairs, creating a new results dataframe called 'high_corr'
(Note: AI was used in parts to assist in editing/modifying the code to its final form below)
"""
corr_abs = corr.abs() #1.
mask = np.triu(np.ones_like(corr_abs, dtype=bool)) #2. & 3.
filtered_corr = corr_abs.mask(mask).stack().reset_index() #4.
filtered_corr.columns = ['Feature_List1', 'Feature_List2', 'Correlation'] #5.
high_corr = ( #6.
    filtered_corr[filtered_corr['Correlation'] > 0.9]
    .query('Feature_List1 != Feature_List2').reset_index(drop=True)
)
```

```python
high_corr.head(10) #checking if the above code works
```

```python
k = 10
col = 'Close'  # Target column
if col in corr.columns:
    feature_corr = corr[col].drop(col).abs().sort_values(ascending=False)
    top_features = feature_corr.head(k)
    print(f"Top {k} features by |correlation| with {col}:\n")
    print(top_features)
else:
    print(f"'{col}' not in columns: {list(corr.columns)}")
```

```python
#Top 10 Numeric Features by Absolute Correlation with Loan Status() Horizontal Bar Chart)
plt.figure(figsize=(10, 6))

top_features.sort_values().plot(kind='barh')

plt.title(f"Top {k} Numeric Features by Absolute Correlation with ETF Close Price",
          fontsize=14, fontweight='bold')
plt.xlabel("Absolute Correlation")
plt.ylabel("Feature Name")

plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
df['Price_Movement'] = df['Close'] - df['Open']
df['Price_Movement_Pct'] = ((df['Close'] - df['Open'])/df['Open']) * 100
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Day_of_Year'] = df['Date'].dt.dayofyear
df['Day_of_Week'] = df['Date'].dt.dayofweek
df['quarter'] = df['Date'].dt.quarter
```

```python
from pandas.api.types import CategoricalDtype

df = df.assign(
            Price_Movement_Category=lambda x: x["Price_Movement"].apply(
                lambda m: "high" if m > 0 else ("low" if m < 0 else "same")
                )
        )

movement_type = CategoricalDtype(categories=['high','same', 'low'], ordered=True)
df["Price_Movement_Category"] = df["Price_Movement_Category"].astype(movement_type)
df["Price_Movement_Code"] = df["Price_Movement_Category"].cat.codes
```
