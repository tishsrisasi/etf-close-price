```python
fig, axes = plt.subplots(14,4,figsize=(20,30)) #creates a figure and 64 sub-plots

axes=axes.ravel() #flattens 2D array

i=0 #initialises the sub-plot counter

for name,df_group in zip(df_names,df_all): #pairing the above names (df_names) vs counter-part (df_all)
    
    if 'Series' in str(type(df_group)): #for single columns (there's only one to deal with here - commodities)
        
        axes[i].hist(df_group.dropna(), #Drops any NaNs found
                     bins=30, #30 bars
                     alpha=.8, #transparency level (set at 80%)
                     color='teal') #bar colours
        axes[i].set_title(name) #sets just title to feature name
        i+=1 #onto the next sub-plot by adding 1 to the counter
    
    else:
        
        for col in df_group.columns:  # looping through all multi-column
            if i<len(axes): #checks if it exceeds 48 sub-plots 
                axes[i].hist(df_group[col].dropna(),
                             bins=30, #as above
                             alpha=.8, #as above
                             color='blue') #as above
                axes[i].set_title(f'{name}: {col}') #sets group title:group member column here
                i+=1 #as above

plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64","float64","int32","float32"]).columns.tolist()
corr_df = df[numeric_cols].corr(method='pearson')
corr_df
```

```python
mask = np.triu(np.ones_like(corr_df, dtype=bool))

plt.figure(figsize=(12, 9))
sns.heatmap(
    corr_df,
    mask=mask,
    annot=False,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Matrix - Numeric Features", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Find pairs of features with high absolute correlation

# Flatten the correlation matrix into pairs
corr_pairs = (
    corr_df.where(np.triu(np.ones(corr_df.shape), k=1).astype(bool))).stack().reset_index()
corr_pairs.columns = ["feature_1", "feature_2", "correlation"]

# Filter pairs with |corr| above a threshold
threshold = 0.9
high_corr_pairs = corr_pairs[abs(corr_pairs["correlation"]) >= threshold] \
                      .sort_values(by="correlation", ascending=False)

print(f"Highly correlated feature pairs (|corr| >= {threshold}):\n")
print(high_corr_pairs.head(20))
```

```python
k = 10
col = 'Close'  # Target column
if col in corr_df.columns:
    feature_corr = corr_df[col].drop(col).abs().sort_values(ascending=False)
    top_features = feature_corr.head(k)
    print(f"Top {k} features by |correlation| with {col}:\n")
    print(top_features)
else:
    print(f"'{col}' not in columns: {list(corr_df.columns)}")
```

```python
#Top 10 Numeric Features by Absolute Correlation with Loan Status() Horizontal Bar Chart)
plt.figure(figsize=(10, 6))

top_features.sort_values().plot(kind='barh')

plt.title(f"Top {k} Numeric Features by Absolute Correlation with ETF Close Price",
          fontsize=14, fontweight='bold')
plt.xlabel("Absolute Correlation")
plt.ylabel("Feature Name")

plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#Adding features to enhance the dataframe - to see if we get better correlations
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Day_of_Year'] = df['Date'].dt.dayofyear
df['Day_of_Week'] = df['Date'].dt.dayofweek
df['quarter'] = df['Date'].dt.quarter
```

```python
df['Month_Sin'] = np.sin(2 * np.pi * df['Month'] / 12).round(6)
df['Month_Cos'] = np.cos(2 * np.pi * df['Month'] / 12).round(6)

df['Dayofweek_Sin'] = np.sin(2 * np.pi * df['Day_of_Week'] / 7)
df['Dayofweek_Cos'] = np.cos(2 * np.pi * df['Day_of_Week'] / 7)
```

```python
#adding some lag elements - as its a time series data
lags = [1, 3, 5, 7, 14, 21, 28, 365]
for lag in lags:
    df[f'Close_lag{lag}'] = df['Close'].shift(lag)

df.fillna(0, inplace=True)
```

```python
# finally, adding some new features based on existing ones
df['Price_Movement'] = df['Close'] - df['Open']
df['Price_Movement_Pct'] = ((df['Close'] - df['Open'])/df['Open']) * 100
df['Price_Range'] = df['High'] - df['Low']
```

```python
#new target variable based on Price Movement plus machine learning considerations
from pandas.api.types import CategoricalDtype

df = df.assign(
            Price_Movement_Category=lambda x: x["Price_Movement"].apply(
                lambda m: "high" if m > 0 else ("low" if m < 0 else "same")
                )
        )

movement_type = CategoricalDtype(categories=['high','same', 'low'], ordered=True)
df["Price_Movement_Category"] = df["Price_Movement_Category"].astype(movement_type)
df["Price_Movement_Code"] = df["Price_Movement_Category"].cat.codes
```
